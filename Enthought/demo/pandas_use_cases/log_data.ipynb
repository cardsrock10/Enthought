### Merging datasets together

```
dataset2 = dataset2.reset_index().set_index("DEPTH")
```

```
# Let's make a fake second dataset...
vs1 = dataset2["VS"]
vs2 = vs1 + randn(len(dataset2["VS"]))*3
#... which contains less data
vs2 = vs2[::2]
print vs2
```

```
# if we do operations on datasets they will align automatically
vs = (vs1 + vs2)/2.
print vs
# or you can align them manually (more efficient if you will do more)
vs2, vs1 = vs2.align(vs1)
# let's force some values in vs1 to be missing
vs1[::5] = nan
# Interpolate the nans away
vs1 = vs1.interpolate()
vs2 = vs2.combine_first(vs1)
```

```
print vs
vs2.plot()
```

```
df = pd.DataFrame({"Original":vs1, "Modified":vs2})
df.plot()
```

### Group-by operations

```
# Let's group the data by lithology and compute some average values for the properties
grp = dataset2.groupby("LITH")
print grp.groups
```

```
# We can apply the same custom operation across each column for each group
grp.aggregate(mean)
```

```
# Or different aggregation functions for a few columns
grp.agg({"VP":min, "VS": max})
```

```
# The group by doesn't have to be on a column name, it can also be a custom function on the index
grp2 = dataset1.groupby(lambda x: int(x/100))
print grp2.groups.keys()
```

```
# Let's compute the average value for each 100 meter sections
# mean, min, max, and a few others are directly available as methods on the groupby object
grp2.mean()
```

```
# Instead of reducing, let's transform each group by removing the average over each group
transformed_data = grp.transform(lambda x: x-x.mean())
```

```
transformed_data
```

```
data_store = pd.HDFStore("datasets.h5")
data_store["transformed"] = transformed_data
data_store["initial"] = dataset1
data_store["without_na"] = dataset2
data_store.close()
```
