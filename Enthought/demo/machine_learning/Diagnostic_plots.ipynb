```
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

matplotlib.rc('font', size=18)
matplotlib.rc('figure', figsize=(10.0, 6.0))
np.random.seed(73473)
```

```
x = stats.norm.rvs(size=50)
y = stats.gamma.rvs(4.0, 0.5, size=50)
_ = plt.hist(x)
plt.show()
_ = plt.hist(y)
```

```
matplotlib.rc('figure', figsize=(6.0, 6.0))
(osm, osr), (slope, intercept, r) = stats.probplot(x, dist='norm', plot=plt)
plt.xlim([-3, 3])
plt.ylim([-3, 3])
print 'Slope =', slope
print 'Intercept =', intercept
print 'r^2 =', r**2
# Plot y=x diagonal line.
plt.plot([-3, 3], [-3, 3], 'k:')
plt.grid(True)
```

```
probplot = sm.ProbPlot(y, stats.gamma, distargs=(4.0,), loc=0.5)
fig = probplot.qqplot(line='45')
plt.grid()
```

```
probplot = sm.ProbPlot(y, stats.gamma, fit=True)
fig = probplot.qqplot(line='45')
```

```
pp_x = sm.ProbPlot(x, fit=True)
pp_y = sm.ProbPlot(y, fit=True)
fig = pp_x.qqplot(line='45', other=pp_y)
plt.grid()
```

```
pp_x = sm.ProbPlot(x)
pp_y = sm.ProbPlot(y)
fig = pp_x.ppplot(line='45', other=pp_y)
```

```
np.random.seed(73473)

matplotlib.rc('figure', figsize=(6.0, 6.0))
font = {'size'   : 18}
matplotlib.rc('font', **font)

matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 

N = 100
x = stats.norm.rvs(loc=0.0, scale=1.0, size=N)
probplot = sm.ProbPlot(x, stats.norm, loc=0.0, scale=1.0)
fig = probplot.qqplot(line='45')
plt.grid()
plt.show()
fig = probplot.ppplot(line='45')
plt.grid()
```

```
probplot = sm.ProbPlot(x * 2, stats.norm, loc=0.0, scale=1.0)
fig = probplot.qqplot(line='45')
plt.grid()
plt.show()
fig = probplot.ppplot(line='45')
plt.grid()
```

```
probplot = sm.ProbPlot(x + 2, stats.norm, loc=0.0, scale=1.0)
fig = probplot.qqplot(line='45')
plt.grid()
plt.show()
fig = probplot.ppplot(line='45')
plt.grid()
```

```
x = stats.t.rvs(2.5, size=N)
print x.mean(), x.std()
probplot = sm.ProbPlot(x, stats.norm, loc=0.0, scale=1.0)
fig = probplot.qqplot(line='45')
plt.grid()
plt.show()
fig = probplot.ppplot(line='45')
plt.grid()
```

```
# Empirical gamma vs theoretical normal
np.random.seed(17373)

y = stats.gamma.rvs(4.0, 0.5, size=100)
probplot = sm.ProbPlot(y, stats.norm, fit=True)
fig = probplot.qqplot(line='45')
plt.grid()
ax = fig.axes[0]
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```
fig = probplot.ppplot(line='45')
plt.grid()
ax = fig.axes[0]
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```
from scipy import stats
import statsmodels.api as sm

# Generate some example data.
x = stats.norm.rvs(size=50,
        loc=2.3, scale=1.5)

# Create Q-Q plot
probplot = sm.ProbPlot(x, stats.norm,
                       fit=True)
probplot.qqplot(line='45')
```
