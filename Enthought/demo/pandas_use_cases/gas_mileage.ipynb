```
import pandas as pd
from matplotlib.pyplot import xlabel, ylabel, title, grid
%matplotlib inline
```

```
# Set some display options
pd.set_option('precision', 3)
```

```
# Read in the data file
datapath = 'civic_gas_mileage.csv'
dta = pd.read_csv(datapath, parse_dates='Date', index_col='Date')
dta
```

```
dta['Gallons']
```

```
# Compute some additional columns
dta['Miles'] = dta['Odometer'].diff()
dta['MPG'] = dta['Miles'] / dta['Gallons']
dta['DPG'] = dta['Price'] / dta['Gallons']
dta['DPM'] = dta['Price'] / dta['Miles']
```

```
dta
```

```
dta['MPG'].dropna().plot(kind='kde')
xlabel('MPG')
title('Estimated PDF of MPG')
```

```
dta.plot(x='Odometer', y='MPG', style='-o')
ylabel('MPG')
xlabel('Mileage')
title("MPG vs Mileage")
grid(True)
```

```
dta.plot(y='MPG', style='-o', grid=True)
ylabel("MPG")
title("MPG vs Date")
```

```
dta.plot(y='DPM', style='-o', grid=True)
ylabel("Dollars/Mile")
title("Dollars / Mile vs Date")
```
